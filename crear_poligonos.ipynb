```python
import pandas as pd
import json
import os

# Leer el archivo CSV
file_path = 'csv_files/data_collars_in_spain.csv'  # Cambia esto si el archivo está en otro lugar
data = pd.read_csv(file_path)

# Crear un archivo JSON por cada id_user
for id_user, group in data.groupby('id_user'):
    # Obtener las coordenadas de lat y lng
    coordinates = group[['lng', 'lat']].values.tolist()
    
    # Crear la estructura del JSON
    geojson = {
        "type": "Feature",
        "properties": {},
        "geometry": {
            "type": "Polygon",
            "coordinates": [coordinates]
        }
    }
    
    output_dir = 'json_poligonos_nuevos'
    json_file_path = os.path.join(output_dir, f"{id_user}.json")
    
    # Guardar el JSON en un archivo
    with open(json_file_path, 'w') as json_file:
        json.dump(geojson, json_file, indent=4)
    
    print(f"Archivo {json_file_path} generado exitosamente.")
```

```python
import os
import json
import folium
import geopandas as gpd
from shapely.geometry import shape
from matplotlib import cm
import random

# Ruta a la carpeta con los archivos JSON generados
ruta_carpeta = './json_poligonos_nuevos'  

# Ruta al shapefile del mapa político de España
ruta_shapefile = './shp_files/ll_autonomicas_inspire_peninbal_etrs89.shp' 

# Cargar el mapa político de España
mapa_espana = gpd.read_file(ruta_shapefile)
print(f"CRS del mapa político antes de la transformación: {mapa_espana.crs}")

# Definir CRS si no está presente
if mapa_espana.crs is None:
    mapa_espana = mapa_espana.set_crs("EPSG:25830")  

# Transformar el CRS al requerido por Folium
mapa_espana = mapa_espana.to_crs("EPSG:4326")
print(f"CRS del mapa político después de la transformación: {mapa_espana.crs}")

# Crear un mapa centrado en España
mapa = folium.Map(location=[40, -3], zoom_start=6)

# Añadir el mapa político de España como GeoJson
folium.GeoJson(data=mapa_espana, name="Mapa político de España").add_to(mapa)

# Generar una lista de colores únicos para los polígonos
color_palette = cm.get_cmap('tab10', 10)  # Paleta de colores con 10 colores
colors = [f"#{random.randint(0, 0xFFFFFF):06x}" for _ in range(len(os.listdir(ruta_carpeta)))]  # Generar colores aleatorios

# Iterar sobre todos los archivos en la carpeta
for idx, archivo in enumerate(os.listdir(ruta_carpeta)):
    if archivo.endswith('.json'):  # Verificar que el archivo tenga extensión .json
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        # Leer y procesar el archivo JSON
        with open(ruta_archivo, 'r') as f:
            datos = json.load(f)
            # Obtener la geometría y convertirla a un objeto Shapely
            geometria = shape(datos['geometry'])
            # Añadir la geometría al mapa con un color y un popup para el nombre del archivo
            folium.GeoJson(
                data=geometria,
                style_function=lambda x, color=colors[idx]: {
                    'fillColor': color,
                    'color': color,
                    'weight': 2,
                    'fillOpacity': 0.5,
                },
                tooltip=archivo  # Mostrar el nombre del archivo al pasar el cursor
            ).add_to(mapa)

# Guardar y mostrar el mapa interactivo
mapa.save("mapa_interactivo_poligonos_nuevos.html")
print("Mapa interactivo guardado como 'mapa_interactivo_poligonos.html'. Ábrelo en tu navegador.")
```

Envolvente convexa y Filtrado de outlayers

```python
import os
import json
import pandas as pd
from shapely.geometry import Polygon
from shapely.geometry import MultiPoint
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Ruta a la carpeta con los archivos JSON generados previamente
ruta_carpeta_json = './json_poligonos_nuevos'

# Función para obtener la envolvente convexa de un conjunto de puntos
def obtener_envolvente_convexa(coordenadas):
    # Crear un objeto MultiPoint con las coordenadas
    puntos = MultiPoint(coordenadas)
    # Obtener la envolvente convexa (el polígono más pequeño que rodea los puntos)
    envolvente = puntos.convex_hull
    return envolvente

# Función para procesar cada archivo JSON y detectar outliers con Isolation Forest
def procesar_json_archivo(ruta_archivo):
    with open(ruta_archivo, 'r') as f:
        # Leer el archivo JSON
        datos = json.load(f)
        
        # Obtener las coordenadas del polígono
        coordenadas = datos['geometry']['coordinates'][0]
        
        # Crear un DataFrame de pandas con las coordenadas
        df = pd.DataFrame(coordenadas, columns=["lng", "lat"])
        
        # Obtener la envolvente convexa (puntos extremos)
        envolvente_convexa = obtener_envolvente_convexa(coordenadas)
        
        # Obtener las coordenadas de los puntos extremos (envolvente convexa)
        puntos_extremos = list(envolvente_convexa.exterior.coords)
        
        # Mostrar los puntos extremos
        print(f"Puntos extremos para {ruta_archivo}: {puntos_extremos}")
        
        # Aplicar Isolation Forest para detectar outliers
        df_extremos = pd.DataFrame(puntos_extremos, columns=["lng", "lat"])
        iso_forest = IsolationForest(contamination=0.1)  # Ajusta el parámetro de contaminación si es necesario
        outliers = iso_forest.fit_predict(df_extremos)

        # Filtrar las coordenadas que no son outliers
        coordenadas_filtradas = df_extremos[outliers == 1]

        # Generar el polígono con las coordenadas filtradas
        poligono_filtrado = Polygon(coordenadas_filtradas.values.tolist())
        
        # Mostrar el polígono filtrado
        print(f"Polígono filtrado para {ruta_archivo}: {poligono_filtrado}")
        
        # Guardar el polígono filtrado en un nuevo archivo GeoJSON
        geojson_filtrado = {
            "type": "Feature",
            "properties": {},
            "geometry": {
                "type": "Polygon",
                "coordinates": [coordenadas_filtradas.values.tolist()]
            }
        }
        
        # Generar el nombre del archivo de salida usando el nombre del archivo original
        nombre_archivo_salida = os.path.join('./json_poligonos_nuevos_filtrados', f"filtrado_{os.path.basename(ruta_archivo)}")
        
        # Crear la carpeta de salida si no existe
        if not os.path.exists('./json_poligonos_nuevos_filtrados'):
            os.makedirs('./json_poligonos_nuevos_filtrados')
        
        # Guardar el archivo filtrado
        with open(nombre_archivo_salida, 'w') as f_salida:
            json.dump(geojson_filtrado, f_salida, indent=4)
        
        print(f"Archivo filtrado guardado en {nombre_archivo_salida}")

# Iterar sobre todos los archivos JSON en la carpeta de entrada
for archivo in os.listdir(ruta_carpeta_json):
    if archivo.endswith('.json'):  # Verificar que el archivo tenga extensión .json
        ruta_archivo = os.path.join(ruta_carpeta_json, archivo)
        procesar_json_archivo(ruta_archivo)
```
